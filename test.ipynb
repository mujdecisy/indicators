```python
import yfinance as yf
import diti
```

```python
start_date = diti.Diti("2023-10-01")
start_date.edit().timezone_update(diti.DitiTimezone.EUROPE__ISTANBUL).head_of(diti.DitiParts.DAYS).commit()

end_date = diti.Diti("2023-11-01")
end_date.edit().timezone_update(diti.DitiTimezone.EUROPE__ISTANBUL).tail_of(diti.DitiParts.DAYS).commit()

print(start_date.to_dt())
print(end_date.to_dt())
```

```python
data = yf.download(
    "TRY=X XU100.IS GC=F BTC-USD",
    start=start_date.to_dt(),
    end=end_date.to_dt(),
    interval="1h",
    auto_adjust=True,
)
data
```
